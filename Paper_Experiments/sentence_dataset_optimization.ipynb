Pipeline for paragraph data based finetuned bert-for-patent

```python
from transformers import pipeline

classifier = pipeline("text-classification", "fassahat/anferico-bert-for-patents-finetuned-150k-patent-paragraphs", framework="tf")
```

```python
import pandas as pd

sentence_data = pd.read_csv('../Dataset/tmp/sentence_based_data_940K.csv')
sentences = sentence_data['sentence'].to_list()
```

```python
classifier_result = classifier(sentences[:1000])
classifier(sentences[0])
```

```python
# classifier_result
selected_sentences = []
labels = []
for i in range(len(classifier_result)):
    if (classifier_result[i]['score'] >= 0.7):
        selected_sentences.append(sentences[i])
        labels.append(classifier_result[i]['label'])
        # print(classifier_result[i]['score'])

final_data = {'sentence': selected_sentences, 'label': labels}
df_result = pd.DataFrame(final_data)
df_result.to_csv('paragraph_model_predicted_sentences.csv', index=False)
```

```python
# classifier_result
classifier_new = pipeline("text-classification", "fassahat/anferico-bert-for-patents-finetuned-150k-patent-paragraphs", framework="tf")
results = []
for i in range(0, len(sentences[:100])):
    results += classifier_new(sentences[i:i+8], labels)

results
```
